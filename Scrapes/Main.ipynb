```python
import requests

standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
data = requests.get(standings_url)
data.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(data.text)
standings_stable = soup.select('table.stats_table')[0]
print(standings_stable)

```

```python
links = standings_stable.find_all('a')
links = [l.get("href") for l in links]
links = [l for l in links if '/squads/' in l]

links
```

```python
team_urls = [f"https://fbref.com{l}" for l in links]
team_urls
```

```python
team_url = team_urls[0]
data0 = requests.get(team_url)
```

```python
import pandas as pd
import requests

matches = pd.read_html(data0.text, match="Scores & Fixtures")
matches[0]
```
